```python
ratings = ratings[ratings['user_id'].isin(y)]
```

```python
# Now let's merge ratings with books

ratings_with_books = ratings.merge(books, on='ISBN')
```

```python
ratings_with_books.head(2)
```

```python
ratings_with_books.shape
```

```python
number_rating = ratings_with_books.groupby('title')['rating'].count().reset_index()
```

```python
number_rating.rename(columns = {
    "rating":'num_of_rating'
}, inplace=True)
```

```python
number_rating.head(2)
```

```python
final_rating =ratings_with_books.merge(number_rating, on ="title")
```

```python
final_rating.head()
```

```python
# Let's take those books which got at least 50 ratings of user
final_rating =final_rating[final_rating['num_of_rating']>=50]
```

```python
final_rating.head(2)
```

```python
# Let's drop the duplicates
final_rating.drop_duplicates(['user_id','title'], inplace=True)
final_rating.isnull().sum()
```

```python
# Lets create a pivot table

book_pivot = final_rating.pivot_table(columns='user_id', index='title', values='rating')
```

```python
book_pivot
```

```python
book_pivot.shape
```

```python
book_pivot.fillna(0, inplace=True)
```

```python
book_pivot
```

```python
from scipy.sparse import csr_matrix
```

```python
book_sparse = csr_matrix(book_pivot)
```

```python
book_sparse
```

```python

type(book_sparse)
```

```python
from sklearn.neighbors import NearestNeighbors
model = NearestNeighbors(algorithm = 'brute')
```

```python
model.fit(book_sparse)
```

```python
book_pivot.iloc[25,:]
```

```python
distance, suggestion = model.kneighbors(book_pivot.iloc[25,:].values.reshape(1,-1), n_neighbors=5)
```

# Testing Model

```python
def recommend_book(book_name):
    book_id = np.where(book_pivot.index == book_name)[0][0]
    distance, suggestion = model.kneighbors(book_pivot.iloc[book_id,:].values.reshape(1,-1), n_neighbors=6)
    
    for i in range(len(suggestion)):
        books = book_pivot.index[suggestion[i]]
        for j in books:
            if j == book_name:
                print(f"Your searched `{book_name}`\n")
                print("The suggestion books are:\n")
            else:
                print(j)
```

```python
book_name = "The Poisonwood Bible: A Novel"
recommend_book(book_name)
```
